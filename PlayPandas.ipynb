```python
def catergorize_athelete(row): #Function to categorize athlete based on height and weight
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] < 80:
        return 'Mediumweight'
    else:
        return 'Heavyweight'
    
bios['athlete_category'] = bios.apply(catergorize_athelete, axis=1) #Apply function to each row, 1= rows, 0=columns
bios[['name', 'height_cm', 'weight_kg', 'athlete_category']]
```

```python



# Total units sold per product in each region using pivot table
pd.pivot_table( 
    df,
    values="Units",
    index="Product",
    columns="Region",
    aggfunc="sum"
)
```

```python
df["Revenue"] = df["Units"] * df["UnitPrice"]
# Total revenue per store in each category using pivot table
pd.pivot_table(
    df,
    values="Revenue",
    index="Store",
    columns="Category",
    aggfunc="sum"
)
```

```python
# Average revenue per product per region using pivot table
pd.pivot_table(
    df,
    values="Units",
    index="Date",
    columns="Product",
    aggfunc="mean"
)
```

```python
# Multi-level pivot table showing total revenue per product per store per region
pd.pivot_table(
    df,
    values="Revenue",
    index=["Region", "Store"],
    columns="Product",
    aggfunc="sum",
    fill_value=0
)
```
